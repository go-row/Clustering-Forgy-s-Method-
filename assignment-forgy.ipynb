<h1>Forgy's Algorithm</h1>

<h3>Importing libraries</h3>

```python
from sklearn.datasets import make_blobs
import numpy as np
from matplotlib import pyplot as plt
from math import sqrt as sq
%matplotlib inline
import time
```

<h3>Generating datapoints</h3>

```python
#x = np.random.rand(5000)
#y = np.random.rand(5000)                       
a, b = make_blobs(n_samples=5000, centers=3, n_features=2)
x = a[:,1]
y = a[:,0]
plt.scatter(x, y, s = 0.3)
plt.show()
```

```python
        #a[:,1].shape
```

<h3>Initial Seeds</h3>

```python
seed1 = [x[50], y[100]]
seed2 = [x[1000], y[2000]]
seed3 = [x[3000], y[200]]
```

<h3>List for storing and coloring clusters</h3>

```python
show1x = []
show1y = []
show2x = []
show2y = []
show3x = []
show3y = []
cluster = np.zeros(5000, dtype = np.int8)
```

<h3>Forgy's Method</h3>

```python
l1 = 1
l2 = 1
l3 = 1
count = [0, 0, 0, 0]
start = time.time();
while(True):
    flag = 0
    changes = 0
    for i in range(0, 5000):
        distance1 = sq((x[i] - seed1[0])**2 + (y[i] - seed1[1])**2)
        distance2 = sq((x[i] - seed2[0])**2 + (y[i] - seed2[1])**2)
        distance3 = sq((x[i] - seed3[0])**2 + (y[i] - seed3[1])**2)
        minimum = min(distance1, distance2, distance3)
        if(minimum == distance1):
            if(cluster[i] != 1):
                count[cluster[i]] -= 1
                flag = 1
                changes += 1
                if(cluster[i] == 0):
                        print("Point", i, "moved to cluster", 1)
                else:
                        print("Point", i, "moved from", cluster[i], "to cluster", 1)
                cluster[i] = 1
                count[cluster[i]] += 1
        elif(minimum == distance2):
            if(cluster[i] != 2):
                count[cluster[i]] -= 1
                flag = 1
                changes += 1
                if(cluster[i] == 0):
                        print("Point", i, "moved to cluster", 2)
                else:
                        print("Point", i, "moved from", cluster[i], "to cluster", 2)
                cluster[i] = 2
                count[cluster[i]] += 1
        else:
            if(cluster[i] != 3):
                count[cluster[i]] -= 1
                flag = 1
                changes += 1
                if(cluster[i] == 0):
                        print("Point", i, "moved to cluster", 3)
                else:
                        print("Point", i, "moved from", cluster[i], "to cluster", 3)
                cluster[i] = 3
                count[cluster[i]] += 1
                
    print("Total sample points which changed their clusters", changes)
    #print("Samples in clusters 1, 2, 3 respectively : ", count[1], count[2], count[3])
    for i in range(0,5000):
        tt = cluster[i]
        if tt == 1:
            seed1[0] = (seed1[0]*l1 + x[i])/(l1+1)
            seed1[1] = (seed1[1]*l1 + y[i])/(l1+1)
            l1 += 1
        elif tt == 2:
            seed2[0] = (seed2[0]*l2 + x[i])/(l2+1)
            seed2[1] = (seed2[1]*l2 + y[i])/(l2+1)
            l2 += 1
        else:
            seed3[0] = (seed3[0]*l3 + x[i])/(l3+1)
            seed3[1] = (seed3[1]*l3 + y[i])/(l3+1)
            l3 += 1
    if flag == 0:
        break
print("Time taken in Forgy's Algorithm : ", time.time()-start)
```

<h3>Plotting clusters</h3>

```python
for i in range(0, 5000):
    if(cluster[i] == 1):
        show1x.append(x[i])
        show1y.append(y[i])
    elif(cluster[i] == 2):
        show2x.append(x[i])
        show2y.append(y[i])
    else:
        show3x.append(x[i])
        show3y.append(y[i])
plt.scatter(show1x, show1y, c = 'red', s = 1)
plt.scatter(show2x, show2y, c = 'yellow', s = 1)
plt.scatter(show3x, show3y, c = 'green', s = 1)
plt.scatter(seed1[0], seed1[1], c = 'black')
plt.scatter(seed2[0], seed2[1], c = 'black')
plt.scatter(seed3[0], seed3[1], c = 'black')
```
